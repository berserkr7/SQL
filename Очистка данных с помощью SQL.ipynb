# Задача 1

**Напишите запрос, который выбирает апрельские данные без повторений осадков. Если в дни D1 и D2 выпало N осадков, то в результаты должен попасть только один из этих дней (более поздний).**

```python
sql = '''

with data as (
select t.*,
row_number() over w as rownum
from weather2 as t
where t.wdate between '2020-04-01' and '2020-04-30'
window w as (
partition by t.precip
order by t.wdate desc)
)


select t.wdate, t.wtemp, t.precip from data as t
where t.rownum = 1
order by t.wdate
'''
select(sql)
```

# Задача 2

**Для некоторых дней не указаны осадки (precip). Заполните их как среднее арифметическое значение осадков от предыдущего и следующего дней. Включите в ответ дни с 6 по 11 марта и с 1 по 6 июня.**

```python
sql = '''
select t.wdate, t.precip,
coalesce(t.precip, avg(t.precip) over w) as fixed
from weather2 as t
where (t.wdate between '2020-03-06' and '2020-03-11')
or (t.wdate between '2020-06-01' and '2020-06-06')
window w as (
order by t.wdate
rows between 1 preceding and 1 following
)
order by t.wdate

'''
select(sql)
```

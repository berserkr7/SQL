# Задача 1

**Есть таблица продаж sales. Посчитайте выручку для тарифа gold по месяцам 2020 года.**

**Для каждого месяца дополнительно укажите:**

* выручку за предыдущий месяц (prev);
* процент, который составляет выручка текущего месяца от prev (perc).

```python
sql = '''
select t.year, t.month, t.revenue,
lag(t.revenue, 1) over w as prev,
round(t.revenue*100.0 / (lag(t.revenue, 1) over w), 0) as perc
from sales as t
where t.year = 2020 and t.plan = 'gold'
window w as (
partition by t.year
rows between unbounded preceding and unbounded following
)
'''
select(sql)
```

# Задача 2

**Есть таблица продаж sales. Посчитайте выручку нарастающим итогом по каждому тарифному плану за первые три месяца 2020 года:**

```python
sql = '''
select t.plan, t.year, t.month, t.revenue,
sum(t.revenue) over w as total
from sales as t
where t.year = 2020 and t.month <= 3
window w as (
partition by t.plan
order by t.month
)
order by t.plan, t.month

'''
select(sql)
```

# Задача 3

**Есть таблица продаж sales. Посчитайте скользящую среднюю выручку за 3 месяца для тарифа platinum в 2020 году.**

```python
sql = '''
select t.year, t.month, t.revenue,
avg(t.revenue) over w as avg3m
from sales as t
where t.year = 2020 and t.plan = 'platinum'
window w as (
partition by t.year
rows between 1 preceding and 1 following
)
order by t.month

'''
select(sql)
```

# Задача 4

**Есть таблица продаж sales. Посчитайте выручку по месяцам для тарифа silver.**

**Для каждого месяца дополнительно укажите:**

* выручку за декабрь этого же года (december);
* процент, который составляет выручка текущего месяца от december (perc).

```python
sql = '''
select t.year, t.month, t.revenue,
last_value(t.revenue) over w as december,
round((t.revenue * 100.0 / last_value(t.revenue) over w) , 0) as perc
from sales as t
where t.plan = 'silver'
window w as (
partition by t.year
order by t.month
rows between unbounded preceding and unbounded following
)
order by year, t.month
'''
select(sql)
```

# Задача 5

**Есть таблица продаж sales. Посчитайте, какой вклад (в процентах) внес каждый из тарифов в общую выручку за год.**

```python
sql = '''

with data as (
select t.year, t.plan, sum(t.revenue) as revenue from sales as t
group by t.plan, t.year
order by t.year, t.plan

)

select t.*,
sum(t.revenue) over w as total,
round(t.revenue * 100.0 / sum(t.revenue) over w , 0) as perc
from data as t
window w as (
partition by t.year)

'''
select(sql)
```

# Задача 6

**Есть таблица продаж sales. Разбейте месяцы 2020 года на три группы по выручке:**

* tile = 1 — высокая,
* tile = 2 — средняя,
* tile = 3 — низкая.

```python
sql = '''
with data as (
select t.year, t.month,
sum(t.revenue) as revenue
from sales as t
where t.year = 2020
group by t.month
)

select t.*,
ntile(3) over w as tile
from data as t
window w as (
order by t.revenue desc
)
'''
select(sql)
```

# Задача 7

**Есть таблица продаж sales. Посчитайте выручку по кварталам 2020 года.**

**Для каждого квартала дополнительно укажите:**

* выручку за аналогичный квартал 2019 года (prev);
* процент, который составляет выручка текущего квартала от prev (perc).

```python
sql = '''

with data as (
select t.year, t.quarter,
sum(t.revenue) as revenue
from sales as t
group by t.year, t.quarter
),

data_1 as (

select t.*,
lag(t.revenue, 4) over w as prev,
round(t.revenue*100.0 / lag(t.revenue, 4) over w, 0) as perc
from data as t
window w as (
order by t.year, t.quarter)
)

select t.* from data_1 as t
where t.prev is not null
order by t.quarter

'''
select(sql)
```

# Задача 8

**Есть таблица продаж sales. Составьте рейтинг месяцев 2020 года с точки зрения количества продаж (quantity) по каждому из тарифов. Чем больше подписок тарифа P было продано в месяц M, тем выше место M в рейтинге по тарифу P. Например, в декабре было продано больше подписок silver, чем в любой другой месяц (1-е место в столбце silver). Для тарифа gold декабрь на 9-м месте, для platinum — на 6-м. Январь же оказался самым слабым месяцем для всех трех тарифов.**

```python
sql = '''


with data as (

select t.year, t.month,
sum(case when t.plan = 'silver' then t.quantity end) as silver,
sum(case when t.plan = 'gold' then t.quantity end) as gold,
sum(case when t.plan = 'platinum' then t.quantity end) as platinum

from sales as t
where t.year = 2020
group by t.month
order by t.month

)


select t.year, t.month,
rank() over(order by t.silver desc) as silver,
rank() over(order by t.gold desc ) as gold,
rank() over(order by t.platinum desc) as platinum
from data as t
order by t.month

'''
select(sql)
```
